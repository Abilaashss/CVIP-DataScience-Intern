```python
df=df[['iyear','imonth','iday','country_txt','region_txt','gname','provstate','city','latitude','longitude','location','summary',
      'attacktype1_txt','targtype1_txt','weaptype1_txt','nkill','nwound','property']]
```

```python
Attacks=df['iyear'].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={'iyear':'years'})
```

```python
Attacks.head() 
```

```python
#Timeline of attack over the years
plt.plot(Attacks['years'], Attacks['count'])
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Number of Attacks Over the Years')
plt.show()
```

```python
#Total killed in each Year
Killed_annual=df[['iyear','nkill']].groupby('iyear').sum()
Killed_annual.head()
```

```python
#Total wounded in each year
Wounded_annual=df[['iyear','nwound']].groupby('iyear').sum()
Wounded_annual.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

#Visualisation-Killed in each year
Killed_annual.plot(kind="bar",color="red",figsize=(15,14),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Year")
ax0.set_ylabel("Number of People Killed") 

#Visualisation-Wounded in each year
Wounded_annual.plot(kind="bar",color="blue",figsize=(15,14),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

```python
#Terrorist Attack distribution across the regions
Terr_region=pd.crosstab(df.iyear,df.region_txt)
Terr_region.head(10)
```

```python
#Visualisation-Region-wise attack
Terr_region.plot(kind="area", stacked=False, alpha=0.7,figsize=(20,10))
plt.title("Region wise attack",fontsize=18)
plt.xlabel("Year",fontsize=18)
plt.ylabel("Number of Attacks",fontsize=18)
plt.show()
```

```python
region_transpose=Terr_region.transpose()
region_transpose["Total"]=region_transpose.sum(axis=1)
ra=region_transpose["Total"].sort_values(ascending=False)
ra
```

```python
#Visualisation-Total Attack in each region
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each region")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
#Killed in each region
Kill_region=df[["region_txt","nkill"]].groupby("region_txt").sum().sort_values(by="nkill",ascending=False)
Kill_region

```

```python
#Wounded in each region
Wounded_region=df[["region_txt","nwound"]].groupby("region_txt").sum().sort_values(by="nwound",ascending=False)
Wounded_region
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#Visualisation-Killed in each region 
Kill_region.plot(kind="bar",color="red",figsize=(15,8),ax=ax0)
ax0.set_title("People Killed in each Region")
ax0.set_xlabel("Region")
ax0.set_ylabel("Number of People Killed")

#Visualisation-Wounded in each region
Wounded_region.plot(kind="bar",color="blue",figsize=(15,8),ax=ax1)
ax1.set_title("People Wounded in each Region")
ax1.set_xlabel("Region")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

```python
#Country-wise attack
Terr_country=df["country_txt"].value_counts().head()
Terr_country
```

```python
#Visualisation-Country wise attack
Terr_country.plot(kind="bar",color="red",figsize=(15,6))
plt.title("Country wise Attacks",fontsize=15)
plt.xlabel("Countries",fontsize=15)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

```python
#Killed in each country
Killed_country=df[["country_txt","nkill"]].groupby("country_txt").sum().sort_values(by="nkill",ascending=False)
Killed_country.head(10)
```

```python
#Wounded in each country
Wounded_country=df[["country_txt","nwound"]].groupby("country_txt").sum().sort_values(by="nwound",ascending=False)
Wounded_country.head(10)
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#Visualisation-Killed in each country 
Killed_country[:10].plot(kind="bar",color="red",figsize=(15,8),ax=ax0)
ax0.set_title("People Killed in each Country")
ax0.set_xlabel("Country")
ax0.set_ylabel("Number of People Killed")

#Visualisation-Wounded in each country 
Wounded_country[:10].plot(kind="bar",color="blue",figsize=(15,8),ax=ax1)
ax1.set_title("People Wounded in each Country")
ax1.set_xlabel("Country")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

```python
#Killed in each city
Killed_city=df[["city","nkill"]].groupby("city").sum().sort_values(by="nkill",ascending=False).drop("Unknown")
Killed_city.head(10)
```

```python
#Wounded in each city
Wounded_city=df[["city","nwound"]].groupby("city").sum().sort_values(by="nwound",ascending=False).drop("Unknown")
Wounded_city.head(10)
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#Visualisation-Killed in each city
Killed_city[:10].plot(kind="bar",color="red",figsize=(15,8),ax=ax0)
ax0.set_title("People Killed in each City")
ax0.set_xlabel("City")
ax0.set_ylabel("Number of People Killed")

#Wounded in each city visualization
Wounded_city[:10].plot(kind="bar",color="blue",figsize=(15,8),ax=ax1)
ax1.set_title("People Wounded in each City")
ax1.set_xlabel("City")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```
